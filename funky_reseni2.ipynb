```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("./vzorky/poisson_intensity_0.2/700.png")
```

```python
# Pevod obrzku do edotnu
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Metoda prmrovn
blur = cv2.blur(gray_image, (5, 5))

# Metoda medinovho filtru
median = cv2.medianBlur(gray_image, 5)

# Metoda Gaussova filtru
gaussian = cv2.GaussianBlur(gray_image, (5, 5), 0)

# Zobrazen pvodnho a filtrovanho obrzku
plt.subplot(2, 2, 1), plt.imshow(gray_image, cmap='gray')
plt.title('Pvodn obrzek'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 2), plt.imshow(blur, cmap='gray')
plt.title('Prmrovn'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 3), plt.imshow(median, cmap='gray')
plt.title('Medinov filtr'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 4), plt.imshow(gaussian, cmap='gray')
plt.title('Gaussovsk filtr'), plt.xticks([]), plt.yticks([])

plt.show()
```
